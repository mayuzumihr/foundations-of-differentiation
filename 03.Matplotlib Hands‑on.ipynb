Ex.01：以下のプログラムを作成してください

1. 数式 $y=sin(x)$ を実装した関数 func_01() を作成する
1. 数式 $y=sin(x) * \pi$ を実装した関数 func_02() を作成する
1. 数式 $y=sin(x) * \pi * 2$ を実装した関数 func_03() を作成する
1. x 軸のデータ列を 0 ～ 10 の範囲で 100 個生成する
1. func_01() のグラフと以下の条件で作成する
    - 線：実線
    - 色：青
1. func_02() のグラフと以下の条件で作成する
    - 線：点線
    - 色：緑
1. func_03() のグラフと以下の条件で作成する
    - 線：破線
    - 色：オレンジ
1. 上記のグラフを重ねて出力する
1. それぞれにタイトル、ラベル、凡例、グリッド、範囲を設定すること
1. x 軸のメモリを 2 ずつとすること

```python
def func_01(x):
    return np.sin(x)

def func_02(x):
    return np.sin(x) * np.pi

def func_03(x):
    return np.sin(x) * np.pi * 2

x = np.linspace(0, 10, 100)
y1 = func_01(x)
y2 = func_02(x)
y3 = func_03(x)

plt.plot(x, y1, color="blue", label="$sin(x)$")
plt.plot(x, y2, color="green", label="$sin(x) * \pi$")
plt.plot(x, y3, color="orange", label="$sin(x) * \pi * 2$")

plt.title("Plot Sample")
plt.xlabel("X")
plt.ylabel("y")

plt.legend()
plt.grid()
plt.show()
```

Ex.02：以下のプログラムを作成してください

1. 以下のランダムな値をそれぞれ x 軸、y 軸とした 2D の散布図を作成する
    - x = np.random.normal(0, 1, 500)
    - y = np.random.normal(0, 1, 500)
1. 色は赤系とする
1. それぞれに適切にタイトル、ラベル、凡例、グリッド、範囲を設定すること

```python
x = np.random.normal(0, 1, 500)
y = np.random.normal(0, 1, 500)

plt.scatter(x, y, color="darkorange", label="Normal Distribution")

plt.title("Scatter Sample")
plt.xlabel("X")
plt.ylabel("y")
plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.legend()
plt.grid()
plt.show()
```

Ex.03：以下のプログラムを作成してください

1. 以下のランダムな値をそれぞれ x 軸、y 軸とした 3D の散布図を作成する
    - x = np.random.normal(0, 1, 500)
    - y = np.random.normal(0, 1, 500)
    - z = np.random.normal(0, 1, 500)
1. 色は青系とする
1. それぞれにタイトル、ラベル、グリッド、範囲を設定すること

```python
#%matplotlib widget

x = np.random.normal(0, 1, 500)
y = np.random.normal(0, 1, 500)
z = np.random.normal(0, 1, 500)

fig = plt.figure()                            # 描画領域の準備              
ax = fig.add_subplot(projection="3d")         # 3D としてグラフを追加
ax.scatter(x, y, z, color="dodgerblue", label="Normal Distribution")    # 凡例、色、マーカー記号、マーカーサイズを指定

ax.set_title("Scatter 3D Sample")     # タイトルの設定
ax.set_xlabel("Label_X")              # X 軸のラベルを指定
ax.set_ylabel("Label_Y")              # Y 軸のラベルを指定
ax.set_zlabel("Label_Z")              # Z 軸のラベルを指定
ax.set_xlim(-3, 3)                    # X 軸の範囲を指定
ax.set_ylim(-3, 3)                    # Y 軸の範囲を指定
ax.set_zlim(-3, 3)                    # Z 軸の範囲を指定

plt.legend()
plt.show()
```
